```python
!ls
from langchain_core.documents import Document
from ragbase.model import create_reranker, create_llm
from langchain.retrievers.document_compressors.chain_filter import LLMChainFilter
from langchain.globals import set_verbose

set_verbose(True)
```

```python
reranker = create_reranker()
```

```python
documents = [
    Document("""Not since Mercedes-Benz introduced the W210 E-Class in 1995 has there been
a bigger transformation that will surprise more Mercedes purists than the full
electrification of the S-Class. Traditionalists, however, are not necessarily
buyers, and those with the budget won't mind the cutting-edge technology at all,
because if Mercedes didn't announce this new EQS was fully electric, no one would
guess by how it performs on the road.
"""),
    Document("""For now, the new 108.4-kWh battery pack operating with exceptional efficiency,
its dual electric motor setup, and advanced regenerative braking system are
featured in the EQS 580, which previously would have been powered by a
high-performance version of the twin-turbo V8 found in the S 560.
"""),
    Document("""
The Hyperscreen display is revolutionary, and Mercedes has done impressive work
with its interface. There are multiple different display modes, including a
classic analog-style cluster, but the most remarkable is an efficiency-focused
mode that shows power consumption and range optimization. Mercedes could have
achieved even greater acclaim if it had maintained the traditional wood trim
options alongside the digital innovation, like the classic interiors of the W126.
"""),
    Document("""The challenging aspect is that the EQS sedan starts at $102,310. The premium
package is no longer standard, though adding the Executive Rear Seat Package is
available as an option. That's significant investment, but until Mercedes sees
demand decrease, you can be certain the brand will continue increasing its
premium positioning. That's fundamental market economics, and many customers
will eagerly embrace this all-electric luxury sedan.
""")
]
```

```python
reranked_documents = reranker.compress_documents(documents, "What is the price?")
```

```python
first_result = reranked_documents[0]
print(first_result.__dict__.keys())
print("score:", first_result.metadata["relevance_score"])
print(first_result.page_content)
```
